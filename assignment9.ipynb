Q4. Custom Tokenization & Regex-based Text Cleaning

Take original text from Question 1.
Write a custom tokenization function that: a. Removes punctuation and special symbols, but keeps contractions (e.g., "isn't" should not be split into "is" and "n't"). b. Handles hyphenated words as a single token (e.g., "state-of-the-art" remains a single token). c. Tokenizes numbers separately but keeps decimal numbers intact (e.g., "3.14" should remain as is).
Use Regex Substitutions (re.sub) to: a. Replace email addresses with '' placeholder. b. Replace URLs with '' placeholder. c. Replace phone numbers (formats: 123-456-7890 or +91 9876543210) with '' placeholder.

```python

import re

text = """I love state-of-the-art technology! It's amazing how quickly AI, machine-learning, and robotics are evolving.
Visit https://example.com for more details or contact me at john.doe123@example.com or +91 9876543210."""

# Step 1: Custom Tokenization Function
def custom_tokenizer(text):
    # Remove punctuation except hyphens, apostrophes (for contractions), and decimals
    text = re.sub(r"[^\w\s\.\-']", '', text)

    # Split into tokens
    tokens = text.split()

    return tokens

# Step 2: Regex Substitution Function
def clean_text(text):
    # Replace email addresses with 
    text = re.sub(r'\b[\w\.-]+@[\w\.-]+\.\w{2,4}\b', '', text)

    # Replace URLs with 
    text = re.sub(r'https?://\S+|www\.\S+', '', text)

    # Replace phone numbers with 
    text = re.sub(r'(\+?\d{1,3}[-\s]?\d{10}|\d{3}-\d{3}-\d{4})', '', text)

    return text

# Step 3: Run the functions
# 3a. First clean the text
cleaned_text = clean_text(text)
print("Cleaned Text:\n", cleaned_text)

# 3b. Then tokenize the cleaned text
tokens = custom_tokenizer(cleaned_text)
print("\nCustom Tokens:\n", tokens)
```
