1)ASK THE USER TO ENTER THE VALUE FOR A MATHEMATICAL FUNCTION.
.GENERATE X VALUES FROM -10 TO 10 USING NP.LINSPACE()
.COMPUTE Y VALUES FOR: Y=M-x^2
                       y=M-sin(x)
.plot both function in a single figure:
use different colors and line styles
.add legend ,grid,and title.

```python
import numpy as np
import matplotlib.pyplot as plt

M=float(input("enter the value fir M:"))
x=np.linspace(-10,10,100)
y1=M*x**2
y2=M-np.sin(x)
plt.figure(figsize=(8,6))
plt.plot(x,y1,label=r'$sy=Mx^2$',color='blue',linestyle='-')
plt.plot(x,y2,label=r'$sy=M-sin(x)$',color='red',linestyle='--')
plt.title("mathematical functions")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

2) CREATE A DATASET OF FIVE SUBJECTS AND SCORES.
CONVERT IT INTO A PANDAS DATAFRAME.
PLOT THE SCORES USING SEABORN BAR PLOT WITH:
..DIFFERENT COLORS FOR EACH BAR.
..ANNOTATIONS ON TOP OF EACH BAR.
..TITLE ,AXIS LABELS, AND GRID.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

data = {
    'Subject': ['Math', 'Science', 'English', 'History', 'Art'],
    'Score': [85, 90, 78, 88, 92]
}

df = pd.DataFrame(data)

sns.set_style("whitegrid")

palette = sns.color_palette("husl", len(df))
ax = sns.barplot(x='Subject', y='Score', data=df, palette=palette)

for i, score in enumerate(df['Score']):
    ax.text(i, score + 1, str(score), ha='center', va='bottom', fontsize=12, fontweight='bold')

plt.title("Subject Scores", fontsize=14, fontweight='bold')
plt.xlabel("Subjects", fontsize=12)
plt.ylabel("Scores", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

3).WRITE A PYTHON SCRIPT TO SELECT YOUR ROLL NO AS SEED FOR NUMPY AND THEN GENERATE A DATASET OF 50 VALUES USING np.random.randn().
create a 2x2 subplot layout displaying: ..line plot showing cumulative sum.
                                        ..scatter plot with random noise.
customize with titles,axis labels,and grids.

```python
import numpy as np
import matplotlib.pyplot as plt

roll_no = 12345  
np.random.seed(roll_no)

data = np.random.randn(50)

fig, axes = plt.subplots(2, 2, figsize=(10, 8))

axes[0, 0].plot(np.cumsum(data), marker='o', linestyle='-', color='b')
axes[0, 0].set_title("Cumulative Sum Line Plot")
axes[0, 0].set_xlabel("Index")
axes[0, 0].set_ylabel("Cumulative Sum")
axes[0, 0].grid(True)

noise = np.random.randn(50) * 0.5 
y_values = data + noise
axes[0, 1].scatter(range(50), y_values, color='r', alpha=0.6)
axes[0, 1].set_title("Scatter Plot with Random Noise")
axes[0, 1].set_xlabel("Index")
axes[0, 1].set_ylabel("Value")
axes[0, 1].grid(True)

axes[1, 0].axis("off")
axes[1, 1].axis("off")

plt.tight_layout()
plt.show()
```

4) DOWNLOAD DATA SET FROM A LINK AND APPLY "SEABORN" LIBRARY TO DO THE FOLLOWING:
1.READ TOTAL PROFIT OF ALL MONTHS AND SHOW IT USING A LINE PLOT.
2.READ ALL PRODUCTS SALES DATA AND SHOW IT USING A MULTIPLE PLOT.
3.PLOT BAR CHART FOR ALL THE FEATURES /ATTRIBUTES.


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset from the correct raw URL
url = "company_sales_data.csv"
df = pd.read_csv(url)

# Ensure correct column names (strip spaces)
df.columns = df.columns.str.strip()

# Line plot for total profit over months
plt.figure(figsize=(8, 5))
sns.lineplot(x=df['Month'], y=df['total_profit'], marker='o', color='blue')
plt.title("Total Profit Over Months")
plt.xlabel("Months")
plt.ylabel("Total Profit")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# Multi-line plot for product sales
plt.figure(figsize=(8, 6))
for column in df.columns[1:-1]:  # Exclude 'Month' and 'Total Profit'
    sns.lineplot(x=df['Month'], y=df[column], label=column)

plt.title("Product Sales Over Months")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.show()

# Bar chart for all attributes
df.set_index("Month").plot(kind='bar', figsize=(10, 6))
plt.title("Sales Data Overview")
plt.xlabel("Months")
plt.ylabel("Values")
plt.grid(axis='y', linestyle='--')
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
data_url = "https://raw.githubusercontent.com/AnjulaMehto/MCA/main/company_sales_data.csv"
df = pd.read_csv(data_url)

# 1. Line plot for total profit per month
plt.figure(figsize=(10, 5))
sns.lineplot(x=df['month_number'], y=df['total_profit'], marker='o', linestyle='-')
plt.title('Total Profit per Month')
plt.xlabel('Month')
plt.ylabel('Total Profit')
plt.grid(True)
plt.show()

# 2. Multiline plot for product sales
df_sales = df.drop(columns=['month_number', 'total_profit'])  # Exclude month and profit columns
plt.figure(figsize=(12, 6))
for column in df_sales.columns:
    sns.lineplot(x=df['month_number'], y=df[column], label=column)
plt.title('Monthly Sales Data for Different Products')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)
plt.show()

# 3. Bar chart for all features
plt.figure(figsize=(12, 6))
df.set_index('month_number').plot(kind='bar', figsize=(12, 6))
plt.title('Sales Data - Bar Chart')
plt.xlabel('Month')
plt.ylabel('Values')
plt.legend(loc='upper left')
plt.grid(axis='y')
plt.show()
```
